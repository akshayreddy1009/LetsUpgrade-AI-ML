## Range() function

Range() is a built-in function in python. Range generates the integer numbers between the given starting integer to the stop integer. It can be used in **FOR LOOP** to generate sequance of iterations generated by range function. You can find a simple program to explain range function below.

**Sample program**

    

```python
for i in range(0,6):
    print(i, end=" ")
    
```

```python
# Q4 : Consider a list of 10 elements of integer values. If the number in the list is divisible by 2, print the
#result as "square of that number minus 2".

list = [10,13,16,32,53,21,39,42,5,6]
for i in list:
    if i%2==0:
        print("result is",i**2-2)

```
